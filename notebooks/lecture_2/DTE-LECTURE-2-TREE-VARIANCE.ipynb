```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn.datasets import make_moons

datasets = []
datasets.append(make_moons(500, noise=0.3, random_state=0x1701a))
datasets.append(make_moons(500, noise=0.3, random_state=0x1701b))

plt.figure(figsize=(10, 5))
for idx in range(2):
    X, y = datasets[idx]
    plt.subplot(1, 2, idx+1)
    plt.scatter(X[:, 0], X[:, 1], cmap = ListedColormap(['#880000', '#000088']), c=y)
    plt.title("Sample Dataset {}: ".format(idx+1))
plt.show()
```

__Decision Tree Classifier:__ Let's fit the decision tree on the datasets. 

```python
### Grab the model ###
from sklearn.tree import DecisionTreeClassifier
import numpy as np

def plot_surface(X,y,max_depth) :
    x_min, x_max = X[:, 0].min() - .1, X[:, 0].max() + .1
    y_min, y_max = X[:, 1].min() - .1, X[:, 1].max() + .1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    X_grid = np.c_[xx.ravel(), yy.ravel()]
    
    # Predict
    model = DecisionTreeClassifier()
    model.fit(X,y)
    z = model.predict(X_grid)
    z = z.reshape(xx.shape)
    
    # Plot
    plt.contourf(xx, yy, z, cmap='RdBu', alpha=.8)
    plt.scatter(X[:,0],X[:,1], cmap = ListedColormap(['#880000', '#000088']), c = y)
    plt.title("Prediction Surface with max_depth: {}".format(max_depth))
    
# Plot them all together
plt.figure(figsize=(20,10))
for d in range(2):
    plt.subplot(1,2,d+1)
    X,y = datasets[d]
    plot_surface(X,y,5)
plt.show()
```

__Extra Tree Classifier:__ A single extra tree classifier with max_depth 10.

```python
### Grab the model ###
from sklearn.ensemble import ExtraTreesClassifier

### Plotting Code ###
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np

def plot_surface(X,y,max_depth,n_estimators) :
    x_min, x_max = X[:, 0].min() - .1, X[:, 0].max() + .1
    y_min, y_max = X[:, 1].min() - .1, X[:, 1].max() + .1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    X_grid = np.c_[xx.ravel(), yy.ravel()]
    
    # Predict
    model = ExtraTreesClassifier(n_estimators = n_estimators, random_state = 8675309,max_depth = max_depth)
    model.fit(X,y)
    z = model.predict(X_grid)
    z = z.reshape(xx.shape)
    
    # Plot
    plt.contourf(xx, yy, z, cmap='RdBu', alpha=.8)
    plt.scatter(X[:,0],X[:,1], cmap = ListedColormap(['#880000', '#000088']), c = y)
    plt.title("Prediction Surface: Depth {}, {} Trees".format(max_depth,n_estimators))
    
# Plot them all together
plt.figure(figsize=(20,10))
for d in range(2) :
    plt.subplot(1,2,d+1)
    X,y = datasets[d]
    plot_surface(X,y,10,1)
plt.show()
```
